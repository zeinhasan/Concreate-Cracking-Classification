# Library

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read Data

```python
# Read Dataq
MobileNet_Performance = pd.read_csv('Rekap Hasil/MobileNet Model Performance 2.csv')
```

```python
# Check Data
MobileNet_Performance.head()
```

# Create Visualization Accuracy

```python
# Find the index of the highest point and the lowest point
highest_index = MobileNet_Performance['Accuracy'].idxmax()
lowest_index = MobileNet_Performance['Accuracy'].idxmin()
# Plotting all points in blue
plt.scatter(MobileNet_Performance['Width Multiplier'], MobileNet_Performance['Accuracy'], color='blue', label='All Points')

# Plotting the highest & lowest point in red with larger size
plt.scatter(MobileNet_Performance.loc[highest_index, 'Width Multiplier'], MobileNet_Performance.loc[highest_index, 'Accuracy'], color='red', s=100, label='Highest Point')
plt.scatter(MobileNet_Performance.loc[lowest_index, 'Width Multiplier'], MobileNet_Performance.loc[lowest_index, 'Accuracy'], color='green', s=100, label='Lowest Point')

# Annotate the highest point with its coordinates
x_highest = MobileNet_Performance.loc[highest_index, 'Width Multiplier']
y_highest = MobileNet_Performance.loc[highest_index, 'Accuracy']
plt.annotate(f'({x_highest}, {y_highest})', (x_highest, y_highest), xytext=(x_highest, y_highest))

# Annotate the lowest point with its coordinates
x_lowest = MobileNet_Performance.loc[lowest_index, 'Width Multiplier']
y_lowest = MobileNet_Performance.loc[lowest_index, 'Accuracy']
plt.annotate(f'({x_lowest}, {y_lowest})', (x_lowest, y_lowest), xytext=(x_lowest, y_lowest))

plt.xlabel('Width Multiplier')
plt.ylabel('Accuracy')
plt.title('Width Multiplier vs Accuracy')
plt.legend()
plt.grid(True)  # Add grid lines
plt.show()
```
