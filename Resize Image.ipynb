# Library

```python
# Importing the libraries
import tensorflow as tf
from tensorflow.keras.preprocessing.image import save_img
import matplotlib.pyplot as plt
```

# Load Image

```python
# Load the image
image_path = 'Dataset/Train/Negative/00003.jpg'
image = tf.io.read_file(image_path)
image = tf.image.decode_image(image, channels=3)
image = tf.cast(image, tf.float32)
```

# Display Original Image

```python
# Display original image
plt.imshow(image.numpy().astype("uint8"))
plt.title("Gambar Asli (227x227)")
plt.axis("on")
plt.show()
```

# Resize Image

```python
# Resize the image
resized_image = tf.image.resize(image, [4, 4])
```

# Display Resized Image

```python
# Display resized image
plt.imshow(resized_image.numpy().astype("uint8"), extent=[0, resized_image.shape[1], 0, resized_image.shape[0]])
plt.title("Gambar Telah Diubah Ukuran (4x4)")
plt.axis("on")
plt.show()
```

# Save Resized Image

```python
# Save the resized image
save_img('Studi Kasus/Preprocessing Resize/00003 - Resized Image (4).jpg', resized_image.numpy())
```
