# Library

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read Data

```python
# Read Excel Data sheet Support Vector Machine
SVM_Model_Performance=pd.read_excel('Rekap Hasil/Hasil Analisis.xlsx', sheet_name='Support Vector Machine')
```

```python
# Check the data
SVM_Model_Performance.head()
```

# Based Radius

## Plot Scatter Num Points vs F1-Score

```python
# Color mapping for different kernels
color_map = {1: 'blue', 2: 'orange', 3: 'green', 4: 'red', 5: 'black'}

# Create subplots for each kernel
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Plot scatter plot for each kernel
for ax, (kernel, group) in zip(axs.flat, SVM_Model_Performance.groupby('Kernel')):
    for radius, sub_group in group.groupby('Radius'):
        ax.scatter(sub_group['Num Points'], sub_group['F1-Score'], color=color_map[radius], label=f'Radius {radius}', s=50)
    ax.set_title(kernel)
    ax.set_xlabel('Num Points')
    ax.set_ylabel('F1-Score')
    ax.grid(True)
    ax.legend()

# Adjust layout
plt.suptitle('Scatter Plot Performa Model Support Vector Machine (F1 Score)', fontsize=16)
plt.tight_layout(rect=[0, 0.05, 1, 1])
plt.show()
```

# Based Num Points

## Plot Scatter Num Points vs F1-Score

```python
# Color mapping for different kernels
color_map = {8: 'blue', 24: 'orange', 48: 'green', 80: 'red', 120: 'black'}

# Create subplots for each kernel
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Plot scatter plot for each kernel
handles, labels = [], []
for ax, (kernel, group) in zip(axs.flat, SVM_Model_Performance.groupby('Kernel')):
    for num_points, sub_group in group.groupby('Num Points'):
        handle = ax.scatter(sub_group['Radius'], sub_group['F1-Score'], color=color_map[num_points], label=f'Num Points {num_points}', s=50)
        if kernel == 'Linear':
            handles.append(handle)
            labels.append(f'Num Points {num_points}')
    ax.set_title(kernel)
    ax.set_xlabel('Radius')
    ax.set_ylabel('F1-Score')
    ax.grid(True)

# Add big title
plt.suptitle('Scatter Plot Performa Model Support Vector Machine (F1 Score)', fontsize=16)

# Create a single legend only for the second subplot
axs[1, 0].legend(handles, labels, loc='lower right', title='Legend', fontsize='small')

# Adjust layout
plt.tight_layout(rect=[0, 0.05, 0.9, 1])  # Adjust top margin for the main title
plt.show()
```
